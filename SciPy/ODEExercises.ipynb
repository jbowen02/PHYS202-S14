#### (a) See what happens to the period of the pendulum as you change the initial angle of release. Plot θ vs t for 5 different initial angles, θ0 , spanning 0 to π. (i.e. 0, π/4 , π/2 , 3π/4 , and π ). Put them all on the same plot.

```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def deriv(theta_current,t):  # return derivatives of the array theta
    g = 9.8 #m/s^2
    ell = 0.5 #m
    theta = theta_current[0]
    omega = theta_current[1]
    dtheta = omega
    domega = -(g/ell)*np.sin(theta)
    return np.array([ dtheta, domega ])
```

```python
g = 9.8
ell = 0.5
th0 = [0., pi/4, pi/2, 3*pi/4, pi]

t = np.arange(0.,10.+0.01,0.01)
omega0 = 0.
theta_init_1 = np.array([th0[0], omega0])
theta_1 = odeint(deriv,theta_init_1,t)

theta_init_2 = np.array([th0[1], omega0])
theta_2 = odeint(deriv,theta_init_2,t)

theta_init_3 = np.array([th0[2], omega0])
theta_3 = odeint(deriv,theta_init_3,t)

theta_init_4 = np.array([th0[3], omega0])
theta_4 = odeint(deriv,theta_init_4,t)

theta_init_5 = np.array([th0[4], omega0])
theta_5 = odeint(deriv,theta_init_5,t)

plt.plot(t,theta_1[:,0],'r',label="Initial Angle = 0")
plt.plot(t,theta_2[:,0],'b',label="Initial Angle = $\pi$/4")
plt.plot(t,theta_3[:,0],'g',label="Initial Angle = $\pi$/2")
plt.plot(t,theta_4[:,0],'black',label="Initial Angle = $\pi$/4")
plt.plot(t,theta_5[:,0],'purple',label="Initial Angle = $\pi$")
plt.xlabel(r'$t (seconds)$',fontsize=20)
plt.ylabel(r'$\theta (rad)$',fontsize=20)
#plt.ylim(-1.,1.5)
plt.grid()
plt.legend(loc="upper left", bbox_to_anchor=(1,1), fontsize=15)
plt.show()
```

#### What happens to the period of the pendulum as the initial angle increases from π/4 to 3π/4?

The period increases as the initial angle increases.

#### What happens to the trajectory when the initial angle is 0 and π ?

When the initial angle is zero, the pendulum does not do anything because it has no angular velocity in the x axis.

When the initial angle is π, the pendulum ball is stright up, and as the time progresses, the ball just falls with no angular velocity in the x direction.
